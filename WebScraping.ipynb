```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
def scrapper_taifa_leo(urls,label):
    list = []
    for url in urls:
        html_page = requests.get(url) # Make a get request to retrieve the page
        soup = BeautifulSoup(html_page.content, 'html.parser') # Pass the page contents to beautiful soup for parsing
        article_container = soup.find('div', class_="news-details-layout1")
        paragraphs = article_container.findAll('p')[1:-2]
        # Extract the text from each <p> tag and concatenate them
        result = ''
        for text in paragraphs:
            result += text.text
        list.append(result)
    df = pd.DataFrame([list]).transpose()
    df.columns = ['text']
    df['label'] = label
    return df
```
